```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

churn=pd.read_csv('telco-customer_churn.csv')
```

```python
categorical=churn.select_dtypes(include=['object'])
numerical=churn.select_dtypes(exclude=['object'],include=['float64'])

categorical=categorical.drop(columns=['customerID','TotalCharges'],axis=1)
```

```python
categorical.info()
```

```python
for col in categorical:
    categorical[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
    
```

```python
for col in numerical:
    plt.figure(figsize=(8, 6))
    sns.histplot(numerical[col], kde=True, bins=20)
    plt.title(f'Histogram for {col}')
    plt.show()
```
