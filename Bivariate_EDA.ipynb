```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn import preprocessing

churn=pd.read_csv('telco-customer_churn.csv')
```

## Pearson and Spearman Correlation Analysis

Label Encode Categorical Variables and Drop IDs

```python
categorical=churn.select_dtypes(include=['object'])
numerical=churn.select_dtypes(exclude=['object'])
```

```python
encoder=preprocessing.LabelEncoder()

for col in categorical:
    categorical[col]=encoder.fit_transform(categorical[col])


categorical.head()
numerical.head()
```

```python
churn_labeled=pd.concat([categorical,numerical], axis=1)
```

```python
pearson=churn_labeled.corr(method='pearson')
spearman=churn_labeled.corr(method='spearman')
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(pearson, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Matrix')
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(spearman, annot=True, cmap='coolwarm')
plt.title('Spearman Correlation Matrix')
plt.show()
```
